Берем все полученные файлы, убираем HTML и возвращаем датасет

```python
import os
import sys 
from bs4 import BeautifulSoup
import json
import re
import datetime
import pandas as pd
import openpyxl
```

```python
abs_path = os.path.abspath(os.getcwd()) + '\\'
dir_path = abs_path + 'finam_news_dataset'
dir_path
```

загружаем файлы новостей по месяцам, и объединяем

```python
data = []
for path in os.listdir(dir_path):
  path = dir_path + '\\' + path
  with open(path, 'r', encoding='utf-8') as f:
    temp = json.load(f)
    for item in temp:
      data.append(item)
```

```python
def replaceTextContent(s):
  s = s.encode().decode().replace('\t', '').replace('\xa0', ' ')
  s = re.sub(r'(\n[ ]*)+', '\n', s)
  return s
```

```python
def parseHTML(item):
  soup = BeautifulSoup(item[3])
  item[3] = replaceTextContent(soup.get_text())
  return item[1:]
```

```python
data_copy = data.copy()
sys.getsizeof(data_copy)
```

```python
data_copy=list(map(lambda item: parseHTML(item), data_copy))
```

```python
def comparator(item):
  return datetime.datetime.strptime(item, '%Y-%m-%d %H:%M') 
```

```python
data_copy = sorted(data_copy, key=lambda item: comparator(item[1]))
```

# Сохранение в XLSX

```python
import xlsxwriter

workbook = xlsxwriter.Workbook('news.xlsx')
worksheet = workbook.add_worksheet()

col = 0
worksheet.write_row(0, col, ['title','datetime','news_body'])
for row, data in enumerate(data_copy):
    worksheet.write_row(row+1, col, data)

workbook.close()
```
